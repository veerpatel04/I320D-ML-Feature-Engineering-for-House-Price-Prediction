First select specific columns and then drop rows which have N/A and check for completeness again. If N/A values are dropped before selecting columns, there will be no rows left in the dataset.

```python
# Drop rows where you see at least one NULL entry
df_1 = df_1.dropna()
```

Create two dataframes housing_features_df and sales_price_df containing the above mentioned columns for features and label respectively.

```python
# Select the Features
housing_features_df = df_1[[ "LotShape", "LotConfig", "Neighborhood", "OverallQual", "OverallCond", "YearRemodAdd", "GrLivArea", "MoSold"]]
housing_features_df.head()
```

LotShape, LotConfig and Neighborhood are categorical-nominal variables.

```python
# First we need to transform the columns to lists
lotshape_list = df_1["LotShape"].to_list()
lotconfig_list = df_1["LotConfig"].to_list()
neighborhood_list = df_1["Neighborhood"].to_list()
```

Convert the columns containing categorical variables to 1D lists and then to 2D list of lists.

```python
# We need to convert the 1-dimensional lists into 2D list of lists
lotshape_list_of_lists = []
for name in lotshape_list:
    lotshape_list_of_lists.append([name])

lotconfig_list_of_lists = []
for name in lotconfig_list:
    lotconfig_list_of_lists.append([name])

neighborhood_list_of_lists = []
for name in neighborhood_list:
    neighborhood_list_of_lists.append([name])
```

Transform and extract the data from these list of lists using one-hot encoding so that the data is presented in an array.

```python
# Transform the data using one hot encoding for all the columns which we changed into 2D list of lists
from sklearn.preprocessing import OneHotEncoder

lotshape_encoder = OneHotEncoder()
lotshape_encoder.fit(lotshape_list_of_lists)
print(f"Unique vocabulary items {len(lotshape_encoder.categories_[0])}\n")
lotshape_transformed = lotshape_encoder.transform(lotshape_list_of_lists)
lotshape_transformed = lotshape_transformed.toarray()
print (lotshape_transformed)

lotconfig_encoder = OneHotEncoder()
lotconfig_encoder.fit(lotconfig_list_of_lists)
print(f"Unique vocabulary items {len(lotconfig_encoder.categories_[0])}\n")
lotconfig_transformed = lotconfig_encoder.transform(lotconfig_list_of_lists)
lotconfig_transformed = lotconfig_transformed.toarray()
print (lotconfig_transformed)

neighborhood_encoder = OneHotEncoder()
neighborhood_encoder.fit(neighborhood_list_of_lists)
print(f"Unique vocabulary items {len(neighborhood_encoder.categories_[0])}\n")
neighborhood_transformed = neighborhood_encoder.transform(neighborhood_list_of_lists)
neighborhood_transformed = neighborhood_transformed.toarray()
print (neighborhood_transformed)
```

```python
# create a dataframe back from the array
lotshape_transformed_df = pd.DataFrame(lotshape_transformed)
lotconfig_transformed_df = pd.DataFrame(lotconfig_transformed)
neighborhood_transformed_df = pd.DataFrame(neighborhood_transformed)
```

From the array, transform the arrays back into dataframes and concatenate them with the original housing_features_df. Then drop theoriginal columns so as to not let them class with the new transformed data. Keep checking the previews to see that the code is actually working.

```python
# Now concatenate this feature back to the original dataframe
housing_features_df.reset_index(drop=True, inplace=True)
lotshape_transformed_df.reset_index(drop=True, inplace=True)
lotconfig_transformed_df.reset_index(drop=True, inplace=True)
neighborhood_transformed_df.reset_index(drop=True, inplace=True)
housing_features_transformed_df = pd.concat([housing_features_df,lotshape_transformed_df,lotconfig_transformed_df, neighborhood_transformed_df], axis=1)
housing_features_transformed_df.shape
```

```python
# Drop the original columns
housing_features_transformed_df = housing_features_transformed_df.drop(columns=["LotShape", "LotConfig", "Neighborhood"], axis=1)
housing_features_transformed_df.head()
```

```python
# We will implement standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
housing_features_transformed_df[
    ['OverallQual',
    'OverallCond',
    'YearRemodAdd',
    'GrLivArea',
     'MoSold']] = scaler.fit_transform(housing_features_transformed_df[['OverallQual', 'OverallCond','YearRemodAdd','GrLivArea', 'MoSold']])
housing_features_transformed_df.head()
```
